## Trend Calculation

```python
# Import the numpy library for numerical operations and statsmodels for trend calculation.
import numpy as np
import statsmodels.api as sm

# Define a function to calculate the trend level of a given variable in a dataset.
def calculate_trend_level(data, variable_name):
    # Create an array representing time from 0 to the length of the data minus one.
    time = np.arange(len(data))
    # Perform ordinary least squares regression with a constant term, fitting a trend to the data.
    trend = sm.OLS(data[variable_name], sm.add_constant(time)).fit().fittedvalues
    # Store the calculated trend level in the data frame with a new column name indicating the trend level of the variable.
    data['trend_' + variable_name+'_level'] = trend

# Define a function to calculate the log trend of a given variable in a dataset.
def calculate_trend(data, variable_name):
    # Calculate the natural log of the variable and store it in a new column in the data frame.
    data['log_' + variable_name] = np.log(data[variable_name])
    # Create an array representing time from 0 to the length of the data minus one.
    time = np.arange(len(data))
    # Perform ordinary least squares regression with a constant term, fitting a log trend to the data.
    trend = sm.OLS(data['log_' + variable_name], sm.add_constant(time)).fit().fittedvalues
    # Store the calculated trend in the data frame with a new column name indicating the log trend of the variable.
    data['trend_' + variable_name] = trend

# Calculate the trend of log real GDP and real GDP per capita for Italy using the defined function.
calculate_trend(df_italy, 'rgdpna')
calculate_trend(df_italy, 'rgdpnapc') # 'rgdpna' refers to national-accounts based real GDP with a 2017 baseline.

# Calculate the trend of log real GDP and real GDP per capita for Costa Rica using the same function as for Italy.
calculate_trend(df_costa_rica, 'rgdpna')
calculate_trend(df_costa_rica, 'rgdpnapc')

# Calculate the trend level of real GDP and real GDP per capita for Italy.
calculate_trend_level(df_italy, 'rgdpna')
calculate_trend_level(df_italy, 'rgdpnapc') # Again, 'rgdpna' refers to national-accounts based real GDP with a 2017 baseline.

# Calculate the trend level of real GDP and real GDP per capita for Costa Rica.
calculate_trend_level(df_costa_rica, 'rgdpna')
calculate_trend_level(df_costa_rica, 'rgdpnapc')
```

## TFP Measure

To measure the Total Factor Productivity (TFP), we used the following production function

$ Y = A \cdot K^\alpha \cdot (hL)^{1-\alpha} $

of which：
- $ Y $ : real GDP (rgdpna),
- $ K $ : capital stock (rnna),
- $ h $ : human capital (hc),
- $ L $ : employment (emp
We first set $ \alpha $ as 0.3. This represents the standard algebraic form,we will also explore the robustness of our result against $ \alpha $.

Hence, A (TFP) can be calculated as the "residual" of other elements in the Solow Model:

$ A = Y / [K^\alpha \cdot (hL)^{1-\alpha}] $

```python
# Set the value of the Cobb-Douglas production function parameter alpha, we are using alpha=0.3 here as suggested by David N.Weil in "Economic growth"
alpha = 0.3

# Calculate the Total Factor Productivity (TFP) for Italy using the Cobb-Douglas production function.
# The calculation incorporates real GDP, capital stock (rnna), human capital (hc), and employment (emp) levels.
df_italy['tfp'] = df_italy['rgdpna'] / (df_italy['rnna'] ** alpha * df_italy['hc'] ** (1 - alpha) * df_italy['emp'] ** (1 - alpha))

# Calculate the Total Factor Productivity (TFP) for Costa Rica in a similar manner to Italy.
df_costa_rica['tfp'] = df_costa_rica['rgdpna'] / (df_costa_rica['rnna'] ** alpha * df_costa_rica['hc'] ** (1 - alpha) * df_costa_rica['emp'] ** (1 - alpha))

# Check the robustness of the TFP result to variations in the alpha parameter by evaluating alternative alpha values.
alternative_alpha = 0.35  # An alternative value for alpha.
# Recalculate TFP for Italy using the alternative alpha value.
df_italy['tfp_alternative'] = df_italy['rgdpna'] / (df_italy['rnna'] ** alternative_alpha * df_italy['hc'] ** (1 - alternative_alpha) * df_italy['emp'] ** (1 - alternative_alpha))
# Recalculate TFP for Costa Rica using the same alternative alpha value.
df_costa_rica['tfp_alternative'] = df_costa_rica['rgdpna'] / (df_costa_rica['rnna'] ** alternative_alpha * df_costa_rica['hc'] ** (1 - alternative_alpha) * df_costa_rica['emp'] ** (1 - alternative_alpha))

alternative_alpha_2 = 0.25  # Another alternative value for alpha.
# Recalculate TFP for both countries using the second alternative alpha value.
df_italy['tfp_alternative_2'] = df_italy['rgdpna'] / (df_italy['rnna'] ** alternative_alpha_2 * df_italy['hc'] ** (1 - alternative_alpha_2) * df_italy['emp'] ** (1 - alternative_alpha_2))
df_costa_rica['tfp_alternative_2'] = df_costa_rica['rgdpna'] / (df_costa_rica['rnna'] ** alternative_alpha_2 * df_costa_rica['hc'] ** (1 - alternative_alpha_2) * df_costa_rica['emp'] ** (1 - alternative_alpha_2))

alternative_alpha_3 = 0.4  # A third alternative value for alpha.
# Recalculate TFP for both countries using the third alternative alpha value.
df_italy['tfp_alternative_3'] = df_italy['rgdpna'] / (df_italy['rnna'] ** alternative_alpha_3  * df_italy['hc'] ** (1 - alternative_alpha_3 ) * df_italy['emp'] ** (1 - alternative_alpha_3 ))
df_costa_rica['tfp_alternative_3'] = df_costa_rica['rgdpna'] / (df_costa_rica['rnna'] ** alternative_alpha_3  * df_costa_rica['hc'] ** (1 - alternative_alpha_3 ) * df_costa_rica['emp'] ** (1 - alternative_alpha_3 ))

# Print the calculated TFP results for Italy and Costa Rica to evaluate the impact of different alpha values on TFP.
print("TFP for Italy:")
print(df_italy[['year', 'tfp']])
print("\nTFP for Costa Rica:")
print(df_costa_rica[['year', 'tfp']])
```

## Growth Accounting

For growth accounting purposes, we used the following transformed model

$ \Delta Y = \alpha \cdot \Delta K + (1-\alpha) \cdot \Delta H + (1-\alpha) \cdot \Delta L $

of which：
- $\Delta Y$ : change in real GDP (rgdpna)，
- $\Delta K$ : change in capital stock (rnna)，
- $\Delta H$ : change in human capital (hc)，
- $\Delta L$ : change in employment (emp)。

# Method 1: Growth Accounting in a yearly basis

```python
# Calculate the rate of change for various economic indicators in Italy to analyze growth dynamics.
df_italy['delta_rgdpna'] = df_italy['rgdpna'].pct_change() * 100  # Calculate the percentage change in real GDP.
df_italy['delta_rnna'] = df_italy['rnna'].pct_change() * 100  # Calculate the percentage change in real capital stock.
df_italy['delta_hc'] = df_italy['hc'].pct_change() * 100  # Calculate the percentage change in human capital index.
df_italy['delta_emp'] = df_italy['emp'].pct_change() * 100  # Calculate the percentage change in employment levels.
df_italy['delta_tfp'] = df_italy['tfp'].pct_change() * 100  # Calculate the percentage change in Total Factor Productivity.
df_italy['delta_ctfp'] = df_italy['ctfp'].pct_change()*100  # Calculate the percentage change in TFP level at current PPPs.
df_italy['delta_cwtfp'] = df_italy['cwtfp'].pct_change()*100  # Calculate the percentage change in Welfare-relevant TFP levels at current PPPs.

# Repeat the rate of change calculations for Costa Rica.
df_costa_rica['delta_rgdpna'] = df_costa_rica['rgdpna'].pct_change() * 100  # Corrected variable name for consistency.
df_costa_rica['delta_rnna'] = df_costa_rica['rnna'].pct_change() * 100
df_costa_rica['delta_hc'] = df_costa_rica['hc'].pct_change() * 100
df_costa_rica['delta_emp'] = df_costa_rica['emp'].pct_change() * 100
df_costa_rica['delta_tfp'] = df_costa_rica['tfp'].pct_change() * 100
df_costa_rica['delta_ctfp'] = df_costa_rica['ctfp'].pct_change()*100
df_costa_rica['delta_cwtfp'] = df_costa_rica['cwtfp'].pct_change()*100

# for growth accounting analysis, we use the given value of alpha=0.3 representing the share of capital in income.
alpha = 0.3


```

## Calculating labor productivity

To construct the series of labor productivity, GDP per worker, the following calculation is used:

$ \text{Labour Productivity} = \frac{\text{Real GDP}}{\text{Employed Workers}} $

```python
# Calculating labor productivity
df_italy['labour_productivity'] = df_italy['rgdpna'] / df_italy['emp']
df_costa_rica['labour_productivity'] = df_costa_rica['rgdpna'] / df_costa_rica['emp']

# Printing the results
print("Labour Productivity for Italy:")
print(df_italy[['year', 'labour_productivity']])
print("\nLabour Productivity for Costa Rica:")
print(df_costa_rica[['year', 'labour_productivity']])
```
